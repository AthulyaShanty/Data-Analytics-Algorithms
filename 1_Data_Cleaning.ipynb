```python
import pandas as pd
```


# Alzheimers data

```python
# Load the CSV file into a DataFrame
alzheimers_data = pd.read_csv('/Users/Student/Library/Mobile Documents/com~apple~CloudDocs/SETU/Sem 2/Data Analytics and Algorithms/Assignment/Implementation/Data-Science-Implementation/Data/alzheimers.csv')
alzheimers_data.head()
```

```python
# Check the number of rows and columns in the DataFrame
num_rows, num_columns = alzheimers_data.shape
num_rows, num_columns
```

```python
# Display basic information about the dataset
alzheimers_data.info()
```

```python
# Display summary statistics of the dataset
alzheimers_data.describe()
```

```python
# Check for missing values
missing_values = alzheimers_data.isnull().sum()
print("Missing values in each column:\n", missing_values)
```

```python
# Check if the data is imbalanced
diagnosis_counts = alzheimers_data['Diagnosis'].value_counts()
print("Diagnosis counts:\n", diagnosis_counts)

# Determine if the data is imbalanced
if diagnosis_counts.max() / diagnosis_counts.min() > 1.5:
    print("The data is imbalanced.")
    print("Imbalanced data:\n", diagnosis_counts)
else:
    print("The data is not imbalanced.")
```

```python
# Check for duplicate rows in the dataset
duplicate_rows = alzheimers_data.duplicated()
num_duplicates = duplicate_rows.sum()

if num_duplicates > 0:
    print("There are duplicates in alzheimers dataset.")
else:
    print("There are no duplicates in alzheimers dataset.")
```

```python
# Save the cleaned data to a CSV file
alzheimers_data.to_csv('alzheimers_cleaned_data.csv', index=False)
```
