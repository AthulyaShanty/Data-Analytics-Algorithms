```python
import pickle
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
# Load the training features
with open('student_train_features.pkl', 'rb') as f:
    train_features = pickle.load(f)

# Load the training target
with open('student_train_target.pkl', 'rb') as f:
    train_target = pickle.load(f)

# Load the test features
with open('student_test_features.pkl', 'rb') as f:
    test_features = pickle.load(f)

# Load the test target
with open('student_test_target.pkl', 'rb') as f:
    test_target = pickle.load(f)

# Load the validation features
with open('student_validation_features.pkl', 'rb') as f:
    validation_features = pickle.load(f)

# Load the validation target
with open('student_validation_target.pkl', 'rb') as f:
    validation_target = pickle.load(f)
```

```python
# Create a linear regression model
model = LinearRegression()

# Train the model using the training data
model.fit(train_features, train_target)

# Predict using the test data
test_predictions = model.predict(test_features)

# Predict using the validation data
validation_predictions = model.predict(validation_features)
```

```python
# Evaluate the model on test data
test_rmse = mean_squared_error(test_target, test_predictions) ** 0.5
test_r2 = r2_score(test_target, test_predictions)

# Evaluate the model on validation data
validation_rmse = mean_squared_error(validation_target, validation_predictions) ** 0.5
validation_r2 = r2_score(validation_target, validation_predictions)

print(f"Test Data - RMSE: {test_rmse}, R^2 Score: {test_r2}")
print(f"Validation Data - RMSE: {validation_rmse}, R^2 Score: {validation_r2}")
```

```python
# Plot actual vs predicted for test data
plt.figure(figsize=(10, 5))
plt.scatter(test_target, test_predictions, alpha=0.5)
plt.plot([test_target.min(), test_target.max()], [test_target.min(), test_target.max()], 'k--', lw=2)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs Predicted - Test Data')
plt.show()

# Plot actual vs predicted for validation data
plt.figure(figsize=(10, 5))
plt.scatter(validation_target, validation_predictions, alpha=0.5)
plt.plot([validation_target.min(), validation_target.max()], [validation_target.min(), validation_target.max()], 'k--', lw=2)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs Predicted - Validation Data')
plt.show()
```
